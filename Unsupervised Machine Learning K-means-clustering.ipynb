```python
# importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
# Standarize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Apply K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)
```

```python
# Get cluster assignments and centroids
cluster_labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
# Visualize the result
plt.figure(figsize=(12,5))

# Before k-means
plt.subplot(121)
plt.scatter(X[:, 0], X[:, 1], c='gray')
plt.title('Before K-means')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal width')
```

```python
# After k-means
plt.subplot(121)
scatter = plt.scatter(X[:, 0], X[:, 1], c=cluster_labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=200, linewidths=3, color='r')
plt.title('After k-means')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.colorbar(scatter)
plt.tight_layout()
plt.show()
```

**What is a good adjusted Rand index score?**
- The adjusted Rand Index (ARI) should be interpreted as follows: ARI >= 0.90 excellent recovery; 0.80 =< ARI < 0.90 good recovery; 0.65 =< ARI < 0.80 moderate recovery; ARI < 0.65 poor recovery.

```python
# Evaluate the clustering
from sklearn.metrics import adjusted_rand_score
print(f"Adjusted Rand Index: {adjusted_rand_score(y, cluster_labels):.2f}")
```
